## $\color{blue}{\text{Task: Fit a linear model to the above German case:}}$
First let's take a look at our data. For this we use matplotlib:


```python
# We use the time as the independent variable and the wages as the dependent variable:
x = wageGer['Year']
y = wageGer['Average Annual Wages']

# Now we give the specifications of the graph: 

plt.figure(figsize=(7,7), dpi=110)
plt.plot(x, y, '*', color='red')
plt.xticks(ticks=np.arange(2000, 2020, 2), fontsize=12, rotation=30, ha='center', va='top')
plt.yticks(ticks=np.arange(39128.538896, 44466.304646, 1000), fontsize=12, rotation=0, ha='right', va='top')
#grid are the lines that form the squares in the graph
plt.grid(linestyle='--', linewidth=1, alpha=0.75) 
plt.tick_params(axis='both',bottom=True, top=False, left=True, right=False, direction='in', which='major', grid_color='black')
plt.title('Year vs Average Annual Wages for Germany', fontsize=12)
plt.show()
```

## $\color{red}{\text{From the graph we can see a linear behaviour of our data, so let's fit a linear model. For this we use LinearRegression that we import from sklearn:}}$

```python
from sklearn.linear_model import LinearRegression

x_a = pd.DataFrame({"A": wageGer['Year']}).to_numpy()
y_a = pd.DataFrame({"A": wageGer['Average Annual Wages']}).to_numpy()

model = LinearRegression().fit(x_a, y_a)
print("Model fitted successfully!")
```

Define by m and by b the slope and the y intercept of the linear model, respectively. This is done by calling .coef_ and .intercept

```python
b = model.intercept_
m = model.coef_
print('The y intercept and the slope are:', b, m)
```

Define the linear model as yfit in the form: y=mx+b

```python
yfit = m*x_a + b
```

Now let's plot the data together with the fitted linear model

```python
plt.figure(figsize=(7,7), dpi=100)
plt.plot(x, y, '^', color='red')
plt.plot(x_a, yfit)
plt.xticks(ticks=np.arange(2000, 2020, 2), fontsize=12, rotation=30, ha='center', va='top')
plt.yticks(ticks=np.arange(39128.538896, 44466.304646, 1000), fontsize=12, rotation=0, ha='right', va='top')
plt.grid(linestyle='--', linewidth=0.5, alpha=0.15)
plt.tick_params(axis='both',bottom=True, top=False, left=True, right=False, direction='in', which='major', grid_color='orange')
plt.title('Year vs Average Annual Wages for Germany with fitted linear model', fontsize=12)
plt.show()
```

How good our model is? Does it work satisfactorily?
To answer this question we use the coefficient of determination ùëÖ2R2: a statistical measure of how well the model approximates the real data points. An ùëÖ2R2 of 1 indicates that the regression predictions perfectly fit the data.
We obtain ùëÖ2R2 with .score() called on model


```python
# Define R^2 as r_sq
r_sq = model.score(x_a, y_a)
print('coefficient of determination:', r_sq)
```

From the above graph and the coefficient of determination we can conclude that a linear model works satisfactorily to predict the future behaviour of Mexico's average annual wages. Of course, there are some limitations.
Let's see the y-values of the fitted model:


```python
y_val = model.predict(x_a)
print('Model values:', y_val, sep='\n')
```

Now let's predict the average annual wages of the next 18 years; that is, from 2019 to 2036:

```python
# Here we sum 19 to x_a:
y_pred = model.predict(x_a+19)
print('Model predictions for the next 18 years:', y_pred, sep='\n')
```

If we want to predict the average annual wage in a specific year we can define our linear model as a function and then evaluate in the desired year:

```python
#Define the model as `lin_model`
def lin_mod(x1):
    return m*x1+b

print('The average annual wage in the year 2036 is:', lin_mod(2036))
```
